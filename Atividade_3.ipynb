```python
#4 Renomeie a coluna Type como CarType em df e substitua o ’.’ nos nomes das
#colunas com ’_’
import numpy as np
import pandas as pd
dados2=pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv')
dados2.rename(columns={"Type":'CarType'},inplace = True)

chave = list(dados2.keys())
tamanho = len(dados2)
dicionario = {}

for i in range(tamanho):
    try:
        nome = chave[i].replace('.','_')
    except:
        continue
    else:
        dicionario[chave[i]]=nome
dados2.rename(columns=dicionario).head()        

```
